#Clustering

```python
#Scaling Features of Data
from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler().fit(X_train)
X_train_norm = norm.transform(X_train)
X_test_norm = norm.transform(X_test)
```

```python
# Determine suitable value of k using elbow method
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
  kmeans.fit(X_train)
  wcss.append(kmeans.inertia_)
```

```python
# plot the curve for different values
plt.plot(range(1, 11), wcss, marker = 'o')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Here the elbow method provides a sharp turn at n_clusters=2.
Hence k=2

```python
from sklearn.metrics import silhouette_score
K = range(2, 8)
fits = []
score = []
for k in K:
  # train the model for current value of k on training data
  model = KMeans(n_clusters = k, random_state = 0, n_init='auto').fit(X_train_norm)
  # append the model to fits
  fits.append(model)
  # Append the silhouette score to scores
  score.append(silhouette_score(X_train_norm, model.labels_, metric='euclidean'))
  print(score)
```

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
kmeans.fit(X_train_norm)
silhouette_score(X_train_norm, kmeans.labels_, metric='euclidean')
```

```python
# Getting cluster labels for the test data
test_cluster_labels = kmeans.predict(X_test_norm)
# Plotting the clusters
plt.figure(figsize=(8, 6))
# Getting cluster centers
cluster_centers = kmeans.cluster_centers_
```

```python
# Plot each cluster for test data
for cluster_label in range(kmeans.n_clusters):
  # Select only the data points that belong to the current cluster in test data
  cluster_data_test = X_test_norm[test_cluster_labels == cluster_label]
  # Plot the data points
  plt.scatter(cluster_data_test[:, 0], cluster_data_test[:, 1], label=f'Test Cluster{cluster_label}', marker='o', alpha=0.5)
# Plotting the centroids
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='o', label='Centroids')
plt.title('KMeans Clustering')
plt.legend()
plt.show()
```
